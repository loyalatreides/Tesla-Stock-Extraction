```python
tcs = yf.Ticker("TCS")
```

```python
tcs_share = tcs.history(period = 'max')
```

```python
infy = yf.Ticker("INFY")
```

```python
infy_share = infy.history(period = "max")
```

```python
infy_share
```

```python
wipro = yf.Ticker('WIPRO.NS')
```

```python
wipro_share = wipro.history(period = "max")
```

```python
wipro_share
```

```python
tcs_share['Open'].plot(label = 'TCS', figsize = (15,7))
infy_share['Open'].plot(label = 'INFY')
wipro_share['Open'].plot(label = 'WIPRO_NS')
plt.title('Stock Prices of TCS, Infosys and Wipro')
plt.legend()
```

```python
tcs_share['Volume'].plot(label = 'TCS', figsize = (15,7))
infy_share['Volume'].plot(label = 'INFY')
wipro_share['Volume'].plot(label = 'WIPRO_NS')
plt.title('Volume of Stock traded')
plt.legend()
```

```python
#Market Capitalisation
tcs_share['MarktCap'] = tcs_share['Volume'] * tcs_share['Open']
infy_share['MarktCap'] = infy_share['Volume'] * infy_share['Open']
wipro_share['MarktCap'] = wipro_share['Volume'] * wipro_share['Open']
tcs_share['MarktCap'].plot(label = 'TCS', figsize = (15,7))
infy_share['MarktCap'].plot(label = 'Infosys')
wipro_share['MarktCap'].plot(label = 'Wipro')
plt.title('Market Cap')
plt.legend()
```

```python
tcs_share['MA50'] = tcs_share['Open'].rolling(50).mean()
tcs_share['MA200'] = tcs_share['Open'].rolling(200).mean()
tcs_share['Open'].plot(figsize = (15,7))
tcs_share['MA50'].plot()
tcs_share['MA200'].plot()
plt.title('50-Days average and 200-Days average for TCS')
plt.legend()
```

```python
wipro_share['MA50'] = wipro_share['Open'].rolling(50).mean()
wipro_share['MA200'] = wipro_share['Open'].rolling(200).mean()
wipro_share['Open'].plot(figsize = (15,7))
wipro_share['MA50'].plot()
wipro_share['MA200'].plot()
plt.title('50-Days average and 200-Days average for Wipro')
plt.legend()
```

```python
infy_share['MA50'] = infy_share['Open'].rolling(50).mean()
infy_share['MA200'] = infy_share['Open'].rolling(200).mean()
infy_share['Open'].plot(figsize = (15,7))
infy_share['MA50'].plot()
infy_share['MA200'].plot()
plt.title('50-Days average and 200-Days average for Infosys')
plt.legend()
```

```python
#Scattered plot matrix
data = pd.concat([tcs_share['Open'], infy_share['Open'], wipro_share['Open']], axis = 1)
data.columns = ['TCSOpen', 'InfosysOpen', 'WiproOpen']
scatter_matrix(data, figsize = (15,15), hist_kwds = {'bins' : 250})
```

```python
#volatility
tcs_share['returns'] = (tcs_share['Close']/tcs_share['Close'].shift(1)) - 1
infy_share['returns'] = (infy_share['Close']/infy_share['Close'].shift(1)) - 1
wipro_share['returns'] = (wipro_share['Close']/wipro_share['Close'].shift(1)) - 1
tcs_share['returns'].hist(bins = 100, label = 'TCS', alpha = 0.5, figsize = (15,7))
infy_share['returns'].hist(bins = 100, label = 'Infosys', alpha = 0.5)
wipro_share['returns'].hist(bins = 100, label = 'Wipro', alpha = 0.5)
plt.legend()
plt.title('Percentage Increase in Stock')
```
